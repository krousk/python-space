```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import matplotlib

from PIL import Image

from wordcloud import WordCloud 
from konlpy.tag import Okt    


from collections import Counter

from IPython.display import set_matplotlib_formats

from ckonlpy.tag import Twitter
```

```python
with open( "메세르의 딸(1_25) 수정.txt", 'r') as f:
    lists = f.readlines()

#lists
```

```python
tagger=Twitter()
tagger.add_dictionary(['예알바','칼레건','요울바','타이런','메세르','켈로스','켈루스','츠질라크'],'Noun')
```

```python
okt = Okt()

morphs = []
 
for sentence in lists:
    morphs.append( tagger.pos( sentence ) )
    
#print( morphs )
```

```python
noun_adj_adv_list=[]
 
for sentence in morphs :
    for word, tag in sentence :
        if tag in [ 'Noun' ]  and ( "려고" not in word ) and ( "처럼" not in word ) and ( "면서" not in word ) and ( "고" not in word ) and ( "때문" not in word ) and ( "까지" not in word ) and ( "거" not in word ) and ( "더" not in word ) and ( "었다" not in word ) and ( "그녀" not in word ) and ( "그" not in word ) and ( "것" not in word ) and ( "내" not in word ) and ( "나" not in word ) and ( "수" not in word ) and ( "게" not in word ) and ( "말" not in word ):
            noun_adj_adv_list.append( word )
            
print( noun_adj_adv_list )
```

```python
count = Counter( noun_adj_adv_list )
words = dict( count.most_common(50) ) #common 안에 집어넣으면 출력숫자 제한 가능.
words
```

```python
matplotlib.rc( 'font',family = 'Malgun Gothic' )
set_matplotlib_formats( 'retina' )
matplotlib.rc( 'axes', unicode_minus = False )
```

```python
wordcloud = WordCloud( font_path = 'C:/Windows/Fonts/malgun.ttf',
                      background_color='black',colormap = "Accent_r",
                      width=1500, height=1000).generate_from_frequencies( words )

plt.imshow( wordcloud )
plt.axis( 'off' )
plt.show()
```
