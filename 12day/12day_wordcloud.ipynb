    3. 또 사용하는 윈도우 O/S가 32비트인지, 64비트인지에 따라서 설치 JPype가 다른데, 윈도우 32비트를 사용하고 있다면 win32를, 윈도우 64비트를 사용하고 있다면 win_amd64를 설치해야 합니다. 예를 들어 파이썬 3.6, 윈도우 64비트를 사용 중이라면 JPype1‑0.6.3‑cp36‑cp36m‑win_amd64.whl를 다운로드합니다.

        > pip install JPype1‑0.6.3‑cp36‑cp36m‑win_amd64.whl

    5. 이제 JPype의 설치가 완료되었다면, KoNLPy를 사용할 준비가 되었습니다.

# 한글 word cloud

### 형태소 분석기, counter import

```python
# KoNlPy의 형태소 분석기( 구. Twitter ) Okt 는 형태소를 뜻함.
from konlpy.tag import Okt    

# 단어 카운팅 빈도 수를 체크하는 것.
from collections import Counter

from IPython.display import set_matplotlib_formats
```

### word cloud용 파일 읽기

```python
with open( "korean.txt", 'r', encoding = 'utf-8' ) as f:
    lists = f.readlines()

lists
```

### 형태소 나누기

```python
okt = Okt()

morphs = []
 
for sentence in lists:
    morphs.append( okt.pos( sentence ) )
    
print( morphs )
```

### 명사, 형용사, 부사만 추출

```python
noun_adj_adv_list=[]
 
for sentence in morphs :
    for word, tag in sentence :
        if tag in [ 'Noun' ] and ( "것" not in word ) and ( "내" not in word ) and ( "나" not in word ) and ( "수" not in word ) and ( "게" not in word ) and ( "말" not in word ):
            noun_adj_adv_list.append( word )
            
print( noun_adj_adv_list )
```

### 빈도수로 정렬

```python
count = Counter( noun_adj_adv_list )
words = dict( count.most_common() )
words
```

### word cloud 생성

#### matplotlib font 설정

```python
# font는 Malgun Gothic , 그래프가 한글일 경우 선명하게 나오지 않아 선명하게 하기('retina')
matplotlib.rc( 'font',family = 'Malgun Gothic' )
set_matplotlib_formats( 'retina' )
matplotlib.rc( 'axes', unicode_minus = False )
```

```python
# font_path는 한글 폰트가 있는 경로 
# background_color는 배경색.  
wordcloud = WordCloud( font_path = 'C:/Windows/Fonts/malgun.ttf',
                      background_color='white',colormap = "Accent_r",
                      width=1500, height=1000).generate_from_frequencies( words )

plt.imshow( wordcloud )
plt.axis( 'off' )
plt.show()
```
