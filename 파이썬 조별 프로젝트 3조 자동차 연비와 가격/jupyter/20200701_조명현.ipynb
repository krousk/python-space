# 주제 : 각 변수들로 사회초년생들이 어느 차량을 구매하면 효율적일지 
#           분석을 통해 확인해본다.

#### 주제 선정 배경

사회초년생들은 사회에 경제활동 및 생활을 하면서 차를 구매하는 경우가 많이 생긴다.

교통수단이용실테조사의 자료에 따르면, 사회생활을 하면서 
차를 구입하는 경우가 70%에 달하는것을 알수 있다. ( 생애 첫자 구매 경험.PNG) 첨부 

그렇다면 차를 구매할때 사회초년생들은 중고차와 신차를 고려해 볼수 있는데, 위 자료를 참고하면
신차 구매가 75%, 중고차 25%를 차지 한다는 것을 알수 있다. ( 생애 첫차 구매 유형.PNG ) 첨부

사회초년생들의 차 구매 목적을 알아보면, 대부분 출퇴근, 여행(레저), 교통수단의 이용 불편 등 
여러가지가 있지만, 위 자료를 참고하면, 출퇴근이 82%로 주 목적임을 알수가 있었다. 
(생애 첫자 구매 목적.PNG )

사회초년생이, 첫차 중에 신차를 구매할 것을 생각한다면, 어느 정도 부담이 될 수 있는데,
그중 가장 많이 고려하는 것이 ( 차량 구매시 고려사항.PNG ) 연비와 가격임을 알수 있다.

그러나 나중에 다시 차를 살때도 연비와 성능을 중요시 한다는 것을 위 자료를 통해서도 알수가 있었다. 
( 향후 차량 구매시 고려사항.PNG ) 첨부

위와 같은 조사를 통하여, 사회초년생의 첫자를 연비와 성능, 가격, 안정성, 디자인 등 여러가지 요소를 
고려하여 구매하는데, 어떠한 차량이 추천하면 좋을지에 대해 생각하여, 이에 따른 
주제를 선정하게 되었다.

## 2. 배기량을 기준으로 경차, 소형차, 중형차, 대형차로 구분한다.

## 3. 국내의 경우 배기량 1000cc 미만은 경차, 1600cc 미만은 소형차, 
## 1600cc 이상 2000cc 미만은 중형차, 2000cc 이상은 대형차로 구분하고 있다.

```python
import pandas as pd
```

```python
data = pd.read_csv( 'C:\workspacePy\Py공부내용들\workspacePy/자동차 에너지소비효율등급 인증 현황_190903..csv', 
                     encoding = 'CP949' )
data
```

# 경차

- 경차는 배기량 1000cc 이하 차량으로 총 127개의 관측치가 나온다. 

```python
data[ data[ '배기량' ] < 1000 ]
```

# 소형차

- 소형차는 배기량 1000~1600cc로 총 258개의 관측치가 나온다. 

```python
data[ ( data[ '배기량' ] > 1000 ) & ( data[ '배기량' ] < 1600 ) ]
```

# 중형차

- 중형차는 배기량 1600~2000cc로 총 604개의 관측치가 나온다.

```python
data[ ( data[ '배기량' ] > 1600 ) & ( data [ '배기량' ] < 2000 ) ]
```

# 대형차

- 대형차는 2000cc 이상의 차로 총 1098의 관측치가 나온다.

```python
data[(data['배기량']>2000)]
```
