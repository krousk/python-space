# 6. EDA

## 6.1 신장 순으로 정렬

```python
df.sort_values( by = 'height', ascending = False )
```

## 6.2 신장별 인원수

```python
height_count = df.height.value_counts()
height_count
```

```python
count = df.height.count()
heights = list( height_count.index )
height_counts = list( height_count.values )

for i in range( len( height_counts ) ):
    print( '신장 [ {:4d}cm ] : {:2d}명( {:5.2f}% )'.format( heights[ i ], height_counts[ i ], height_counts[ i ] / count * 100 ) )
```

## 6.3 평균 신장 이상 / 미만 대통령

```python
mean_height_count = df.height.mean()

mask_mean_above = df.height >= mean_height_count
mask_mean_under = df.height < mean_height_count

mean_height_above = df.loc[ mask_mean_above ] # 평균 신장 이상
mean_height_under = df.loc[ mask_mean_under ] # 평균 신장 미만
```

### 6.3.1 평균 신장 이상 대통령

```python
mean_height_above
```

```python
print( '평균 신장 {:6.2f}cm 이상 대통령은 {:2d}명'.format( df.height.mean(), mean_height_above[ 'height' ].count() ) )
```

###  6.3.2 평균 신장 미만 대통령

```python
mean_height_under
```

```python
print( '평균 신장 {:6.2f}cm 미만 대통령은 {:2d}명'.format( df.height.mean(), mean_height_under[ 'height' ].count() ) )
```

## 6.3 최고 신장 대통령 정보

```python
mask = df.height >= df.height.max()
max_height = df.loc[ mask ]
max_height
```

```python
for president_info in max_height.values:
    print( '최고 신장 대통령 정보 : {:3d}cm ( {:2d}대 {:<22s} 대통령 )'.format( president_info[ 2 ], 
                                                                                president_info[ 0 ], 
                                                                                president_info[ 1 ] ) )
```

## 6.4 최저 신장 대통령 정보

```python
mask = df.height <= df.height.min()
min_height = df.loc[ mask ]
min_height
```

```python
for president_info in min_height.values:
    print( '최저 신장 대통령 정보 : {:3d}cm ( {:2d}대 {:<22s} 대통령 )'.format( president_info[ 2 ], 
                                                                                president_info[ 0 ], 
                                                                                president_info[ 1 ] ) )
```

# 7. EDA 결과

1. 미국 대통령들의 평균 신장은 179.74cm 입니다.

3. 미국 대통령중 평균 신장 이상의 대통령은 23명이며, 평균 신장 미만의 대통령은 19명 입니다.
4. 미국 대통령중 최고 신장은 193cm이며 ***16대 Abraham Lincoin***과 ***36대 Lyndon B Johnson*** 대통령입니다.
